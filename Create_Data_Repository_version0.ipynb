```python
import nilearn
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import os
import scipy.misc
from PIL import Image
```

```python
M=['WM','GM']
```

```python
#M=['CSF']   
M=['WM','GM']
for m in M:
    PATH='/Volumes/Gradient/Personal/yun/Echobay/Mask/'+m+'/'
    savepath='/Volumes/Gradient/Personal/yun/Echobay/Mask_png/'+m

        
    thr=[0.5,0.6,0.7,0.8]
    #sa=savepath+'_'+thr+'/'
    for t in thr:
        for i in os.listdir(PATH):
              #print(os.path.basename(PATH+i))
        #         print(os.path.splitext(i)[0][0:-4])
            ana=nib.load(PATH+i)
            anat_img_data = ana.get_fdata()
            anat_img_data.shape

            sa=savepath+'_'+str(t)
            if not os.path.exists(sa):
                os.mkdir(sa)
            else:
                print('all done')

            savename=sa+'/'+os.path.splitext(i)[0][:-4]+'.png'
            #print(savename)

            scipy.misc.imsave(savename, np.where(anat_img_data > t, 1, 0))
```

```python
import pandas as pd
BASE='/Volumes/Gradient/Personal/yun/Echobay/Version0/Mask_png/'
for i in os.listdir(BASE):
    II=os.listdir(BASE+i)
   
    FI=[]
    for mm in II:
        im=Image.open(BASE+i+'/'+mm)

        count=0
        for pixel in im.getdata():
            if pixel==255:
                count+=1
            else:
                count+=0

        if not im.getextrema()[1]==0 and count>400:
            FI.append(mm)
        else:
            print('Blank')
    I=pd.DataFrame(FI,columns=['ID'])  
    I.to_csv(BASE+i+'.csv')

        
        
        
```
